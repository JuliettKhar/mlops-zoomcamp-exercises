```python
import pandas as pd
from evidently import Report
from evidently.metrics import (ValueDrift, QuantileValue)
```

```python
import warnings

warnings.filterwarnings(
    "ignore",
    message="invalid value encountered in scalar divide",
    category=RuntimeWarning
)
```

```python
df = pd.read_parquet('data/green_tripdata_2026-01.parquet')
reference_dataset = pd.read_parquet('data/reference.parquet')
df.head()
```

```python
print(f'Q1: rows is {df.shape[0]}')
```

```python
report = Report([
    ValueDrift(column='fare_amount'), 
    QuantileValue(column='fare_amount', quantile=0.5)
])

snapshot = report.run(
    reference_data=reference_dataset,
    current_data=df
)
```

```python
result = snapshot.dict()
result
```

```python
print(f"Q2: Metric chosen: ValueDrift(column=fare_amount) Quantile metric: QuantileValue(column=fare_amount, quantile=0.5)")
```

```python
from evidently import Report
from evidently.metrics import QuantileValue
import json

quantiles = []

for day in range(1, 32):
    current =  df[df.lpep_pickup_datetime.dt.day == day]

    if len(current) == 0:
        continue

    report = Report([
        QuantileValue(
            column='fare_amount',
            quantile=0.5
        )
    ])

    result = report.run(current_data=current)

    data= json.loads(result.json())
    q50 = data['metrics'][0]['value']
    quantiles.append(q50)

print("Q3 max =", max(quantiles))
```
